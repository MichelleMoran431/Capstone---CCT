```python
import pandas as pd
import scipy.stats as stats

# Load your production data DataFrame
file_path = r'C:\Users\User\Desktop\Thesis 2023\Capstone---CCT\Python Working Notebooks\ProductionDataupdated1.csv'
uploaded_df = pd.read_csv(file_path)

# List of INGRED_ID values
INGRED_ID = ['1461896', '1254972', '1031006', '1243269', '1815609', '1196706']

# Create a dictionary to store results
results = {}

# Perform t-test for each INGRED_ID group
for ingred_id in INGRED_ID:
    group_data = uploaded_df[uploaded_df['INGRED_ID'] == ingred_id]['Phase_overrun']
    mean_phase_overrun = group_data.mean()
    t_statistic, p_value = stats.ttest_1samp(group_data, popmean=mean_phase_overrun)
    results[ingred_id] = {'Mean Phase Overrun': mean_phase_overrun,
                          'T-Statistic': t_statistic,
                          'P-Value': p_value}

# Print the results
for ingred_id, result in results.items():
    print(f"INGRED_ID: {ingred_id}")
    print(f"Mean Phase Overrun: {result['Mean Phase Overrun']:.2f}")
    print(f"T-Statistic: {result['T-Statistic']:.2f}")
    print(f"P-Value: {result['P-Value']:.4f}")
    print("-" * 30)
```
