```python
# Data Collection:
# Using the Specify Absolute Path: If the file is located in a different directory, you can specify the absolute path to the file when reading it using pd.read_csv():
import pandas as pd
file_path = r'C:\Users\User\Desktop\Thesis 2023\Capstone---CCT\Python Working Notebooks\ProductionDataupdated1.csv'
ProductionTank = pd.read_csv(file_path)
```

```python
# Define a dictionary of replacements
replacements = {
    'U_23MT01' : '2301',
    'U_26MT02' : '2602',
    'U_22MT01' : '2201'
}

# Replace values in the 'TankName' column
ProductionTank['Tank_1'].replace(replacements, inplace=True)

# Print the modified DataFrame
print(ProductionTank)
```

```python
ProductionTank2 = ProductionTank.replace(np.nan, 0)
```

```python
ProductionTank2.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report


df = pd.DataFrame(ProductionTank2)
# Assuming df is your dataframe, X are your features and y is your target variable
X = df.drop(['Deaeration Phase','Phase_start','Phase_end','INGRED_Name','Instruction_Step','INGRED_ID'],axis=1)
y = df['Deaeration Phase']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train SVM
svm_model = SVC(kernel='linear')  # Starting with linear kernel
svm_model.fit(X_train, y_train)

# Predict
y_pred = svm_model.predict(X_test)

# Evaluate
cm = confusion_matrix(y_test, y_pred)
report = classification_report(y_test, y_pred)

print("Confusion Matrix:")
print(cm)
print("\nClassification Report:")
print(report)
```
